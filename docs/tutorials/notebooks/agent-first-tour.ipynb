## Outline

1. Start from an offline sample session.
2. Inspect a simple deterministic result.
3. Move to the agent-first `ask()` path when credentials are available.
4. Try a simple scenario variation.


```python
import trellis

s = trellis.quickstart()
bond = trellis.sample_bond_10y()
result = s.price(bond)

{
    "clean_price": round(result.clean_price, 6),
    "dv01": round(result.greeks["dv01"], 6),
}
```

Even if you start with `ask()`, you often want to move into an explicit session for scenario work.
